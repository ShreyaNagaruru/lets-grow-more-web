**TASK 1-IMAGE TO PENCIL SKETCH**

**Importing the libraries**

```python
import numpy as np
import pandas as pd
import cv2 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image ,ImageOps
import matplotlib.pylab as plt
```

**Load the image**

```python
input= cv2.imread("german.jpg")
cv2_imshow(input)
cv2.waitKey(0)
```

**Converting the original image into grey scale image**

```python
gry_image = cv2.cvtColor(input, cv2.COLOR_BGR2GRAY)
cv2_imshow(gry_image)
cv2.waitKey(0)
```

**Inverting the image**

```python
invert_image = 255-gry_image
cv2_imshow(invert_image)
cv2.waitKey()
```

```python
#Blurring the image by gaussian filter
blur=cv2.GaussianBlur(invert_image, (31, 31), 0)
```

```python
invert_blur= 255 - blur
pencil_sketch = cv2.divide(gry_image, invert_blur, scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Pencil sketch")
plt.axis('off')
```
